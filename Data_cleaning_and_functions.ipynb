1. What is data cleaning and why is it important in data analysis?
   
   - What are the potential consequences of analyzing unclean or messy data?

   - Explain the common steps involved in cleaning and organizing data.

#Data Cleaning:

  - Data cleaning is the process of identifying and correcting errors in a dataset. It includes removing duplicate records, fixing incorrect values, handling missing data, and standardizing data formats to make the data accurate and consistent.

#Why is Data Cleaning important:

  -  Data Cleaning is important because accurate data leads to reliable analysis. If the data is clean, the results, insights, and decisions based on that data will be correct and trustworthy.

#Consequences of analyzing unclean or messy data:

  - Incorrect or misleading results

  - Wrong business decisions

  - Reduced accuracy of reports and models

  - Loss of time and credibility

#Common steps involved in cleaning and organizing data:

  - Removing duplicate records

  - Handling missing values

  - Correcting incorrect or inconsistent data

  - Standardizing formats (dates, text, numbers)

  - Removing unnecessary data

  - Validating and accuracy

3. Explain the use of text functions such as TRIM, LEFT, RIGHT, MID, and CONCAT in data cleaning.

#Use of Text Functions in Data Cleaning

  - Text functions are commonly used in data cleaning to format, extract, and combine text data so that it becomes consistent and easy to analyze.

#TRIM()

  - The TRIM() function removes extra spaces from text except single spaces between words.

#Example:

  - =TRIM(A1)

#LEFT

  - The LEFT() function extracts a specified number of characters from the beginning of a text string.

#Example:

  - =LEFT(A1,5)

#RIGHT()

  - The RIGHT() function extracts a specified number of characters from the end of a text string.

#Example:

  - =RIGHT(A1,4)


#MID()

  - The MID() function extracts characters from the middle of a text string based on a starting position and length.

#Example:

  - =MID(A1,2,4)

#CONCAT()

  - The CONCAT() function is used to combine multiple text values into one.

#Example:

  - =CONCAT(A1, " ", B1)

#Why these functions are useful in data cleaning:

  - Remove unwanted spaces

  - Extract specific parts of text

  - Combine text fields

  - Standardize text format

  


4. What is the role of data functions like TODAY in managing datasets?

   - Data functions such as TODAY() are used to work with current dates and help in managing, updating, and analyzing time-based data.

#Role of TODAY() Function:

  - The TODAY() function returns the current date.

  - It updates automatically whenever the worksheet is opened or recalculated.

#Why TODAY() is useful in datasets:

  - Helps track deadlines and due dates

  - Calculates age of data or number of days passed

  - Automatically updates reports with the current date

  - Useful for time-based analysis such as pending tasks or remaining days

#Example use case:

  - If a dataset contains a due date, TODAY() can be used to calculate how many days are left or overdue by subtracting the due date from the current date.

#Understanding TODAY() and NOW()

  - Difference between TODAY() and NOW()

   - TODAY() returns only the current date.

   - NOW() returns the current date and current time.

#Example:

 - =TODAY()

 - =NOW()

#Use case:

  - Use TODAY() when only the date is required (reports, deadlines).

  - Use NOW() when both date and time are required (log entries, timestamps).

#Days left until a future date

  - If cell A1 contains the date 2025-06-10, the formula to find how many days are left from today is:

  - =A1-Today()

#Display current date and time using NOW() in format MM/DD/YYYY HH:MM AM/PM

  - =TEXT (NOW(), "MM/DD/YYYY HH:MM AM/PM")

#What happens if a cell contains =TODAY() when the worksheet is reopened the next day?

  - The value will automatically update to the new current date because TODAY() is a volatile function.

#Store a static date (today's date) without changing

  - Use the keyboard shortcut:

  - Windows: Ctrl + ;

  - Mac: Command + ;

  - This insers today's date as a fixed value, not a formula.
